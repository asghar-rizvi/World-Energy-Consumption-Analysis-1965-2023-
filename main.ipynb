# Complete Data Analysis On World Energy Consumption

### Importing Every Neccessary Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

### Loading Data 

```python
data= pd.read_csv("World_Energy_By_Country_And_Region_1965_to_2023.csv")
```

### Analyzing Data

### Cleaning Of Data

```python
data.dropna()
```

### Handling missing Numeric values 

```python
for year in range(1965, 2024):
    data[str(year)] = pd.to_numeric(data[str(year)], errors='coerce')
```

### Data Visualizations

##### FINDING ENERGY CONSUMPTIONS ON BASIS OF COUNTRY

```python
def country_consumption_energy(country):
    cleaned_data=data[data["Country"]== country].drop(columns="Country").T
    cleaned_data.columns=['Consumption']
    plt.figure(figsize=(15, 8))
    plt.plot(cleaned_data.index, cleaned_data['Consumption'], marker='o',color='r')
    plt.grid()
    plt.title(f"Energy Consumption In {country}")
    plt.xlabel("YEAR",fontsize=15)
    plt.ylabel("Consumption",fontsize=15)
    plt.xticks(rotation=90)
    
country_consumption_energy("Canada")
```

##### FINDING ENERGY CONSUMPTION ON YEAR BASIS

```python
def year_consumption_energy(year):
    plt.figure(figsize=(15,5))
    plt.bar(data["Country"].astype(str),data[year])
    plt.title(f"Energy Consumption In Year {year}")
    plt.xlabel("COUNTRIES",fontsize=15)
    plt.ylabel("CONSUMPTION",fontsize=15)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

year_consumption_energy("1992")
```

##### Correlation Using HeatMap

```python
new_data=data.drop(columns="Country").corr() # Dropping string values (it has only country column)
plt.figure(figsize=(10,5))
sns.heatmap(new_data,cmap="coolwarm",annot=False)
plt.title("HeatMap Of Correlation Of Year And Heat Consumption")
plt.show()
```

##### Growth Rate Of Energy Of Each Country

```python
def growth_rate(data,startYear,endYear):
    data["Growth_Rate"]=((data[endYear]-data[startYear])/data[startYear])*100
    return data

new_data=growth_rate(data,"1965","2023")
# getting top 10 countries on basis of growth rate into another data frame
top_countries_growth_year = new_data.sort_values(by="Growth_Rate", ascending=False).head(10)
print(top_countries_growth_year[["Country", "Growth_Rate"]])

plt.figure(figsize=(11,4))
sns.barplot(data=top_countries_growth_year,x="Country",y="Growth_Rate",color="green")
plt.title("Top 10 Countries And Their Growth From Year 1965-2023")
plt.ylabel("Growth Rate (%)",fontsize=15)
plt.xlabel("Country",fontsize=15)
plt.grid(alpha=0.4)
    
plt.show()
```

##### SEPERATING COUNTRIES AND REGIONS IN 2 DIFFERENT DATA FRAMES USING COPY FUNCTION

```python
regions = [
    'Total North America', 'Central America', 'Other Caribbean', 'Other South America',
    'Total S. & Cent. America', 'Other Europe', 'Total Europe', 'Total CIS', 'Other Middle East',
    'Total Middle East', 'Eastern Africa', 'Middle Africa', 'Western Africa', 'Other Northern Africa',
    'Total Africa', 'Other Asia Pacific', 'Total Asia Pacific', 'Total World', 'of which: OECD', 
    '                 Non-OECD', '                 European Union #'
]

# Create a new DataFrame for regions
regions_data = data[data['Country'].isin(regions)].copy()

# Remove region entries from the original DataFrame
countries_data = data[~data['Country'].isin(regions)].copy()
```

##### Making Plot Top 10 Countries And Their Energy Consumption In Required Year

```python
top_countries=countries_data[["Country",'2023']].sort_values(by='2023',ascending=False).head(10)
print(top_countries)
c = (255/255, 140/255, 0/255)
plt.figure(figsize=(10,5))
sns.barplot(data=top_countries,x='Country',y="2023",color="yellow",saturation=0.2)
plt.xlabel("Country",fontsize=15)
plt.ylabel("Energy Consumption",fontsize=15)
plt.xticks(rotation=45,color=c)
plt.yticks(color=c)
plt.title("Top Countries and their Energy Consumption In Year 2023")
plt.show()
```

##### Making Plot Top 10 Regions And Their Energy Consumption In Required Year

```python
top_regions=regions_data[['Country','2023']].sort_values(by='2023',ascending=False).head(10)
print(top_regions)
c = (255/255, 140/255, 0/255)
# c = (0/255, 0/255, 255/255) # Blue
plt.figure(figsize=(10,5))
sns.barplot(data=top_regions,x='Country',y="2023",color="yellow")
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Energy Consumption",fontsize=15)
plt.xticks(rotation=45,color=c)
plt.yticks(color=c)
plt.title("Top Regions and their Energy Consumption In Year 2023")
plt.show()
```
